```python
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

```python
#year
date['year'] = date['date'].dt.year
date.sample(5)
```

```python
#month
date['month'] = date['date'].dt.month
date.sample(5)
```

```python
#day
date['day'] = date['date'].dt.day
date.sample(5)
```

```python
#month name
date['date_month_name'] = date['date'].dt.month_name()
date.head()
```

```python
#day of week
date['dow'] = date['date'].dt.dayofweek
date.head()
```

```python
#day of week name
date['dow_name'] = date['date'].dt.day_name()
date.head()
```

```python
#is weekend
date['is_weekend'] = np.where(date['dow_name'].isin(['Sunday','Saturday']),1,0)
date.head()
```

```python
#week of the year
date['date_week'] = date['date'].dt.isocalendar().week
date.head()
```

```python
#quarter
date['quarter'] = date['date'].dt.quarter
date.head()
```

```python
#semester
date['semester'] = np.where(date['quarter'].isin([1,2]), 1, 2)
date.head()
```

```python
#Extract Time elapsed between dates
import datetime

today = datetime.datetime.today()
today
```

```python
(today - date['date'])
```

```python
(today - date['date']).dt.days
```

```python
#weeks passed
np.round((today -date['date'])/ np.timedelta64(1, 'W'),0)
```
