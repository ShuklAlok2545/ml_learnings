```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
```

```python
df = pd.read_csv('titanic_toy.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().mean()
```

```python
X = df.drop(columns=['Survived'])
y = df['Survived']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
age_mean = X_train['Age'].mean()
age_median = X_train['Age'].median()

fare_mean = X_train['Age'].mean()
fare_median = X_train['Age'].median()
```

```python
X_train['Age_mean'] = X_train['Age'].fillna(age_mean)
X_train['Age_median'] = X_train['Age'].fillna(age_median)

X_train['Fare_mean'] = X_train['Fare'].fillna(fare_mean)
X_train['Fare_median'] = X_train['Fare'].fillna(fare_median)
```

```python
print('Age Variance original wala imputation: ', X_train['Age'].var())
print('Original Age after mean variance: ', X_train['Age_mean'].var())
print('Age Variance after median wala imputation: ', X_train['Age_median'].var())


print('Original Fare variable variance: ', X_train['Fare'].var())
print('Fare Variance after mean wala imputation: ',X_train['Fare_mean'].var())
print('Fare Variance after -median wala imputation: ', X_train['Fare_median'].var())
```

```python
X_train['Age_99'] = X_train['Age'].fillna(99)
X_train['Age_minus1'] = X_train['Age'].fillna(-1)

X_train['Fare_999'] = X_train['Fare'].fillna(999)
X_train['Fare_minus1'] = X_train['Fare'].fillna(-1)
```

```python
print('Original Age variable variance: ', X_train['Age'].var())
print('Age Variance after 99 wala imputation: ', X_train['Age_99'].var())
print('Age Variance after -1 wala imputation: ', X_train['Age_minus1'].var())

print('Original Fare variable variance: ', X_train['Fare'].var())
print('Fare Variance after 999 wala imputation: ', X_train['Fare_999'].var())
print('Fare Variance after -1 wala imputation: ', X_train['Fare_minus1'].var())
```

```python
imputer1 = SimpleImputer(strategy='constant',fill_value=99)
imputer2 = SimpleImputer(strategy='constant',fill_value=999)

imput1 = SimpleImputer(strategy='mean')
imput2 = SimpleImputer(strategy='median')
```

```python
trf = ColumnTransformer([
    ('imputer1',imputer1,['Age']),
    ('imputer2',imputer2,['Fare'])
],remainder='passthrough')
```

```python
trf.fit(X_train)
```

```python
trf2 = ColumnTransformer([
    ('imput1',imput1,['Age']),
    ('imput2',imput2,['Fare'])
],remainder='passthrough')
```
